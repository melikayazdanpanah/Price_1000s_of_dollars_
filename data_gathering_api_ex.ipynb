```python
# url of scrape
sit_url =" https://en.wikipedia.org/wiki/Data_science"
```

```python
#get the page by launch url
page = requests.get(sit_url)
#print out the first 1000 character of the page that we srcaped
page.content[0:1000]
#messy page of HTML
```

use bs4 for takes messy documents parses based on a specified format

```python
#using html parser 
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# now we have a object we call it soup that contain parsed vertion of html
#now we can selwct particular part of web page
#print oot the page title 
print('TITLE : \n') 
titllle =soup.title
titr = titllle.get_text()
print(titr)

#print out the first p-tag
print('\nP_tag:\n')
first_p=soup.find('p')
paragraph = first_p.get_text()
print(paragraph)
```
